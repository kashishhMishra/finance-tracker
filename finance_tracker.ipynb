```python
import pandas as pd
import matplotlib.pyplot as plt

# simple expense data
data = {
    "Category": ["Food", "Travel", "Shopping", "Bills", "Entertainment"],
    "Amount": [2000, 1500, 3000, 2500, 1000]
}

df = pd.DataFrame(data)

print(df)

# PIE CHART
plt.pie(df["Amount"], labels=df["Category"], autopct='%1.1f%%')
plt.title("Expense Distribution")
plt.show()

# BAR GRAPH
plt.bar(df["Category"], df["Amount"])
plt.title("Expenses by Category")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.show()
```
